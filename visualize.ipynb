```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns    


data=pd.read_csv('cleaned_autocentral_data.csv')

print(data.shape)
print(data.describe())
print(data.head())
```

### some informations about our data

```python
plt.figure(figsize = (20, 15))
sns.countplot(y = data.brand)
plt.title("Car companies with their cars", fontsize = 20)
plt.show()
```

```python
#plt.figure(figsize = (20, 15))
sns.countplot(y = data.transmission)
plt.title("Car transmission with their counts")
plt.show()
```

```python


# Create price ranges (adjust bins as you like)
data["price_range"] = pd.cut(
    data["price"],
    bins=[0, 10000, 50000, 100000, 250000, 500000],
    labels=["0-10k", "10k-50k", "50k-100k", "100k-250k", "250k-500k"]
)

# Plot counts of each price range
sns.countplot(y=data["price_range"])
plt.title("Car Price Ranges with Their Counts")
plt.xlabel("Count")
plt.ylabel("Price Range")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Logical mileage bins
data["mileage_range"] = pd.cut(
    data["mileage"],
    bins=[0, 20000, 60000, 120000, 200000, 350000],
    labels=["0-20k", "20k-60k", "60k-120k", "120k-200k", "200k-350k"]
)

# Plot
sns.countplot(y=data["mileage_range"])
plt.title("Mileage Ranges with Their Counts")
plt.xlabel("Count")
plt.ylabel("Mileage Range")
plt.show()
```

```python
sns.regplot(x="mileage", y="price", data=data)
plt.ylim(0, 500000)
plt.xlim(0, 100000)

plt.show()
```

#### conclusion : 
data is very scattered , no obvious pattern regarding mileage/price linearship

```python
sns.regplot(x="cv", y="price", data=data)
plt.ylim(0, 100000)
plt.xlim(0, 50)

plt.show()
```

#### conclusion : 
same for mileage

```python
top20 = data["brand"].value_counts().head(20).index
df_top20 = data[data["brand"].isin(top20)]


import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
sns.boxplot(x="brand", y="price", data=df_top20)
plt.title("Price Distribution of Top 20 Most Frequent Car Brands")
plt.xticks(rotation=45)
plt.xlabel("Brand")
plt.ylabel("Price")
plt.tight_layout()
plt.show()
```
